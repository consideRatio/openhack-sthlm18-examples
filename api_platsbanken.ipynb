# Platsbanken API example

```python
import requests

# NOTE: Use 'accept-language' --- using 'accept' does not work!
headers={"accept-language": "application/json"}
base_url="https://api.arbetsformedlingen.se/af/v0/platsannonser"

# See: https://jobtechdev.se/swagger/index.html#/platsannonser
resp = requests.get(base_url + "/matchning?lanid=1&antalrader=10", headers=headers)
print(f"Status Code: {resp.status_code}, Ok Status: {resp.ok}")
```

```python
json_data = resp.json()
matches = json_data['matchningslista']['matchningdata']
len(matches)
```
